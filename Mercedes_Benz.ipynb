# Perform dimensionality reduction.

```python
x=train
x.head()
```

```python
y=db_train.y
y
```

```python
print(x.shape)
print(y.shape)
```

# Performing PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=.97)
```

```python
pca.fit(x,y)
```

```python
x_trans=pca.fit_transform(x)
print(np.round(x_trans,2))
```

```python
x_trans.shape
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
```

```python
train_x,test_x,train_y,test_y = train_test_split(x_trans,y,test_size=.3,random_state=7)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
xgb_reg = xgb.XGBRegressor()
```

```python
model = xgb_reg.fit(train_x,train_y)
```

```python
print('RMSE = ',sqrt(mean_squared_error(model.predict(test_x),test_y)))
```
